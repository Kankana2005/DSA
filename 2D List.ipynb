# 2D List

### Print all the elements from a matrix


```python
nums = [[5,20,3],
[7,-10,9],
[1,-52,6]]
rows = len(nums)
cols = len(nums[0])
for i in range(0,rows):
    for j in range(0,cols):
        print(nums[i][j],end=" ")
    print()    
```

### Print upper triangle


```python
nums = [[5,10,8],
        [7,6,3],
        [2,1,9]]
rows = len(nums)
cols = len(nums[0])
for i in range(0,rows):
    for j in range(0,cols):
        if j>=i:
            print(nums[i][j],end = " ")
        else:
            print("*",end = " ")
    print()            
```

### Print lower triangle

```python
nums = [[5,10,8],
        [7,6,3],
        [2,1,9]]
rows = len(nums)
cols = len(nums[0])
for i in range(0,rows):
    for j in range(0,cols):
        if j<=i:
            print(nums[i][j],end = " ")
        else:
            print("*",end = " ")
    print()            
```

### print Diagonal

```python
nums = [[5,10,8],
        [7,6,3],
        [2,1,9]]
rows = len(nums)
cols = len(nums[0])
for i in range(0,rows):
    for j in range(0,cols):
        if j==i:
            print(nums[i][j],end = " ")
        else:
            print("*",end = " ")
    print()            
```

### sum of all elements in a 2d list

```python
nums = [[5,20,3],
[7,-10,9],
[1,-52,6]]
rows = len(nums)
cols = len(nums[0])
sum = 0
for i in range(0,rows):
    for j in range(0,cols):
        sum += nums[i][j]
print(sum)      
    
```

### print another diagonal

```python
nums = [[5,10,8],
        [7,6,3],
        [2,1,9]]
rows = len(nums)
cols = len(nums[0])
for i in range(0,rows):
    for j in range(0,cols):
        if i+j == cols-1:
            print(nums[i][j],end = " ")
        else:
            print("*",end = " ")
    print()            
```

### Transpose of a matrix

```python
nums = [[5,10,8],
        [7,6,3]]
rows = len(nums)
cols = len(nums[0])
res = [[0]* rows for i in range(cols)]
for i in range(0,rows):
    for j in range(0,cols):
        res[j][i] = nums[i][j]
        
print(res)         
```

### Set Matrix zeros

```python
def mark_inf(matrix,row,col):
    r = len(matrix)
    c = len(matrix[0])
    for i in range (0,r):
        if matrix[i][col] !=0:
            matrix[i][col] = float("inf")
    for j in range (0,c):
        if matrix[row][j] !=0:
            matrix[row][j] = float("inf") 
def set_zero(matrix):
    r = len(matrix)
    c = len(matrix[0]) 
    for i in range(0,r):
        for j in range(0,c):
            if matrix[i][j] == 0:
                mark_inf(matrix,i,j) 
    for i in range(0,r):
        for j in range(0,c):
            if matrix[i][j] == float("inf"):
                matrix[i][j] = 0
matrix = [[7,9,2,3],[20,8,0,10],[29,0,-10,5],[4,14,6,7]]
set_zero(matrix)
print(matrix)                                         
```

```python
#Optimal solution
def set_zeros(matrix):
    r = len(matrix)
    c = len(matrix[0])
    rowtrack = [0 for _ in range (r)]
    coltrack = [0 for _ in range (c)]
    for i in range (0,r):
        for j in range(0,r):
            if matrix[i][j] == 0:
                rowtrack[i] = -1
                coltrack[j] = -1
    for i in range (0,r):
        for j in range(0,r): 
            if rowtrack[i] == -1 or coltrack[j] == -1:
                matrix[i][j] == 0
                
matrix = [[7,9,2,3],[20,8,0,10],[29,0,-10,5],[4,14,6,7]]
set_zero(matrix)
print(matrix)
            
                       
                
    
```
