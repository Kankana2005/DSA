## Binary Search

### Coding Problem

```python
# iterative solution

def binary_search(nums,target):
    n = len(nums)
    low = 0
    high = n-1
    while low<=high:
        mid = (low+high)//2
        if nums[mid] == target:
            return mid
        
        elif mid > target:
            high = mid - 1
        else:
            low = mid+1 
    return -1

nums = [2,4,6,7,9,11,18,19]
target =  13
print(binary_search(nums,target))

              
        
```

```python
# Recursive method

def bs(nums, low, high):
    if low> high:
        return -1
    target = 13
    mid = (low+high)//2
    if nums[mid] == target:
        return mid
    elif nums[mid] > target:
        return bs(nums,low,mid-1)
    else:
        return bs(nums,mid+1,high)
    
nums = [2,4,6,7,9,11,18,19]
n= len(nums)
print(bs(nums,0,n-1))
    
```

#### Lower bound and upper bound

```python
# lower bound
def lower_bound(nums):
    n = len(nums)
    low = 0
    high = n-1
    lb = n
    target = 1
    while low<= high:
        mid = (low+high)//2
        if nums[mid] >= target:
            lb = mid
            high = mid-1
        else:
            low = mid+1
    return lb

nums = [1,1,1,2,3,3,5,6,7,7,7,9,12,12,13]   
print(lower_bound(nums))         
        
        
```

```python
# upper bound

def upper_bound(nums,target):
    n = len(nums)
    low = 0
    high = n-1
    ub = n
    while low <= high:
        mid = (low+high)//2
        if nums[mid] > target:
            ub = mid
            high = mid-1
        else:
            low = mid+1
    return ub


nums = [1,1,1,2,3,3,5,6,7,7,7,9,12,12,13]  
target = 6 
print(upper_bound(nums,target))             
```

### Search insertion postion

```python
def search_position(nums,target):
    n= len(nums)
    low = 0
    high = n-1
    lb = n
    while low<= high :
        mid = (low+high)//2
        if nums[mid] >= target:
            lb = mid
            high = mid-1
        else:
            low = mid+1
    return lb

nums = [1,2,3,4,8,9,14,15,19,20,21]
target = 20
print(search_position(nums,target))            
```
