## Binary Search

### Coding Problem

```python
# iterative solution

def binary_search(nums,target):
    n = len(nums)
    low = 0
    high = n-1
    while low<=high:
        mid = (low+high)//2
        if nums[mid] == target:
            return mid
        
        elif mid > target:
            high = mid - 1
        else:
            low = mid+1 
    return -1

nums = [2,4,6,7,9,11,18,19]
target =  13
print(binary_search(nums,target))

              
        
```

```python
# Recursive method

def bs(nums, low, high):
    if low> high:
        return -1
    target = 13
    mid = (low+high)//2
    if nums[mid] == target:
        return mid
    elif nums[mid] > target:
        return bs(nums,low,mid-1)
    else:
        return bs(nums,mid+1,high)
    
nums = [2,4,6,7,9,11,18,19]
n= len(nums)
print(bs(nums,0,n-1))
    
```
