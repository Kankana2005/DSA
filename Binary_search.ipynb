## Binary Search

### Coding Problem

```python
# iterative solution

def binary_search(nums,target):
    n = len(nums)
    low = 0
    high = n-1
    while low<=high:
        mid = (low+high)//2
        if nums[mid] == target:
            return mid
        
        elif mid > target:
            high = mid - 1
        else:
            low = mid+1 
    return -1

nums = [2,4,6,7,9,11,18,19]
target =  13
print(binary_search(nums,target))

              
        
```

```python
# Recursive method

def bs(nums, low, high):
    if low> high:
        return -1
    target = 13
    mid = (low+high)//2
    if nums[mid] == target:
        return mid
    elif nums[mid] > target:
        return bs(nums,low,mid-1)
    else:
        return bs(nums,mid+1,high)
    
nums = [2,4,6,7,9,11,18,19]
n= len(nums)
print(bs(nums,0,n-1))
    
```

#### Lower bound and upper bound

```python
# lower bound
def lower_bound(nums):
    n = len(nums)
    low = 0
    high = n-1
    lb = n
    target = 1
    while low<= high:
        mid = (low+high)//2
        if nums[mid] >= target:
            lb = mid
            high = mid-1
        else:
            low = mid+1
    return lb

nums = [1,1,1,2,3,3,5,6,7,7,7,9,12,12,13]   
print(lower_bound(nums))         
        
        
```

```python
# upper bound

def upper_bound(nums,target):
    n = len(nums)
    low = 0
    high = n-1
    ub = n
    while low <= high:
        mid = (low+high)//2
        if nums[mid] > target:
            ub = mid
            high = mid-1
        else:
            low = mid+1
    return ub


nums = [1,1,1,2,3,3,5,6,7,7,7,9,12,12,13]  
target = 6 
print(upper_bound(nums,target))             
```

### Search insertion postion

```python
#optimal solution

def search_position(nums,target):
    n= len(nums)
    low = 0
    high = n-1
    lb = n
    while low<= high :
        mid = (low+high)//2
        if nums[mid] >= target:
            lb = mid
            high = mid-1
        else:
            low = mid+1
    return lb

nums = [1,2,3,4,8,9,14,15,19,20,21]
target = 20
print(search_position(nums,target))            
```

### Ceil the floor

```python
# Optimal solution
def ceil_floor(nums,target):
    floor = -1
    ceil = -1
    n = len(nums)
    low = 0
    high = n-1
    while low <= high:
        mid = (low+high)//2
        if nums[mid]== target:
            return [nums[mid],nums[mid]]
        elif nums[mid] > target:
            ceil = nums[mid]
            high = mid-1
        else:
            floor = nums[mid]
            low = mid+1
    return [floor,ceil]

nums = [3,4,4,8,9,9,10,12,12,14,15]
target = 11
print(ceil_floor(nums,target))            
        
```

### Find the first and last occurance of an element

```python

# brute force solution

def first_last(nums,target):
    n = len(nums)
    first = -1
    last = -1

    for i in range(0,n):
            if nums[i] == target:
                if first == -1: 
                    first = i
                last = i
            

    return [first,last] 
nums = [1,2,3,3,3,3,3,5,6,8,9,9,10]
target = 3
print(first_last(nums,target))
    
            
                   
```

### Count the occurances of a number

```python
#brute force solution-1
def count(nums,target):
    countt = 0
    for i in range(len(nums)):
        if nums[i] == target:
            countt+=1
           
    return countt

nums = [1,2,3,3,3,3,3,5,6,8,9,9,10]
target = 3
print(count(nums,target))
        
```

```python
# brute force solution -2

def occurance(nums,target):
    n = len(nums)
    first = -1
    last = -1
    for i in range (0,n):
        if nums[i] == target:
            if first == -1:
                first = i
            last = i
    if first == -1:
        return 0
    return(last - first +1)  

nums = [1,2,3,3,3,3,3,5,6,8,9,9,10]
target = 3
print(count(nums,target))          
```
