### Append ,traversal,Delete

```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
       
class SinglyLinkedList:
    def __init__(self) -> None:
        self.head = None
    
    def append(self,data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node 
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node
            
    def traverse(self):
        if not self.head:
            print("SLL is empty")
        else:
            current = self.head
            while current is not None:
                print(current.val, end = " ")
                current = current.next
            print()
            
    def insert_at(self,val,position):
        new_node = Node(val)
        if position == 0:
            new_node.next = self.head 
            self.head = new_node
            
        else:
            current = self.head
            prev_node = None
            count = 0
            while current is not None and count < position:
                prev_node = current
                current = current.next
                count+=1
            prev_node.next = new_node
            new_node.next = current 
            
    


    def delete(self,val):
        temp = self.head
        if temp.next is not None:
            if temp.val == val:
                self.head = temp.next
                return
            else:
                found = False
                prev = None
                while temp is not None:
                    if temp.val == val:
                        found = True
                        break
                    prev = temp
                    temp = temp.next
                    
                if found:
                    prev.next = temp.next
                    return
                else:
                    print("Node not found")    
                                          
            
sll = SinglyLinkedList()
sll.append(10)
sll.append(20)
sll.append(30)
sll.append(40)
sll.append(1)
sll.delete(30)
sll.traverse()  
sll.insert_at(100,3)
sll.traverse()                                              
        
```

### Q1. Find the middle of linked list 

```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
       
class SinglyLinkedList:
    def __init__(self) -> None:
        self.head = None
        
    def insert(self,val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node        
    
    def middle(self):
        n = 0
        temp = self.head 
        
        #count nodes
        while temp is not None:
            n+=1
            temp = temp.next
            
        temp = self.head
        
        for i in range(0,n//2):
            temp = temp.next
        return temp.val
    
sll = SinglyLinkedList()
values = [5,10,27,3,99]
for v in values:
    sll.insert(v)
    
print("Middle: ", sll.middle())              
               
```

```python
#Optimal solution
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
       
class SinglyLinkedList:
    def __init__(self) -> None:
        self.head = None
        
    def insert(self,val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node        
    
    def middle(self):
        slow = self.head
        fast = self.head
        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next
        return slow.val   
    
sll = SinglyLinkedList()
values = [5,10,27,3,99]
for v in values:
    sll.insert(v)
    
print("Middle: ", sll.middle())
```

### Reverse a Linked List

```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
       
class SinglyLinkedList:
    def __init__(self) -> None:
        self.head = None
        
    def insert(self,val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node 
        
    def reverse(self):
        temp = self.head
        stack = []
        while temp is not None:
            stack.append(temp.val)
            temp = temp.next 
        temp = self.head
        while temp is not None:
            e = stack.pop()
            temp.val = e
            temp = temp.next
            
        return self.head.val               
        
sll = SinglyLinkedList()
values = [5,10,27,3,99]
for v in values:
    sll.insert(v)
    
print("reverse: ", sll.reverse())
    
            
```

```python
# Optimal solution

class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
       
class SinglyLinkedList:
    def __init__(self) -> None:
        self.head = None
        
    def insert(self,val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node 
        
    def reverse(self):
        temp = self.head
        prev = None
        while temp is not None:
            front = temp.next
            temp.next = prev
            prev = temp
            temp = front
        return prev.val 
    
sll = SinglyLinkedList()
values = [5,10,27,3,99]
for v in values:
    sll.insert(v)
    
print("reverse: ", sll.reverse())
           
```

### Linked List cycle

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
        
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        
    def insert(self, val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        
    def LinkedListCycle(self):
        temp = self.head
        my_set = set()
        while temp is not None:
            if temp in my_set:
                return True
            my_set.add(temp)
            temp = temp.next
        return False
    
sll = SinglyLinkedList()
values = [5,9,1,7,6,1,9,2,8,1] 
for v in values:
    sll.insert(v)
    
print(sll.LinkedListCycle())
```

### Odd even linked list(rearrange nodes)

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
        
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        
    def insert(self, val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        
    def OddEven(self):
        odd = self.head
        even = self.head.next
        even_head = even
        while even is not None and even.next is not None:
            odd.next = odd.next.next
            odd = odd.next
            even.next = even.next.next
            even = even.next
            
        odd.next = even_head
        return self.head  
    
sll = SinglyLinkedList()
values = [5,9,1,7,6,1,9,2,8,1] 
for v in values:
    sll.insert(v)
    
print(sll.OddEven().val)          
           
```
